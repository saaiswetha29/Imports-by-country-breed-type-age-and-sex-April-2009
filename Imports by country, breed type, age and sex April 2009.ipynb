```python
data["Age"]=data["Age"].fillna(method="ffill")
data["Breed_type"]=data["Breed_type"].fillna(method="ffill")
```

```python
data.drop(44,axis=0,inplace=True)
```

```python
data.drop(["Unnamed:_12"],axis=1,inplace=True)
```
